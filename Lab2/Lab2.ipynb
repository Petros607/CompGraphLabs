```python
import math
import matplotlib.pyplot as plt
```

```python
def deg_to_rad(angle_degrees):
    return angle_degrees * math.pi / 180


#    L-
def draw_lsystem(axiom, rule, angle, iterations, length):
    pattern = axiom
    for _ in range(iterations):
        new_pattern = ""
        for symbol in pattern:
            if symbol == 'F':
                new_pattern += rule
            else:
                new_pattern += symbol
        pattern = new_pattern

    position = [0, 0]  #   (x, y)
    direction = 0  # (  x )

    x_coords = [position[0]]
    y_coords = [position[1]]

    for symbol in pattern:
        if symbol == 'F':
            new_x = position[0] + length * math.cos(deg_to_rad(direction))
            new_y = position[1] + length * math.sin(deg_to_rad(direction))

            x_coords.append(new_x)
            y_coords.append(new_y)
            position = [new_x, new_y]

        elif symbol == '+':
            direction -= angle

        elif symbol == '-':
            direction += angle

    plt.clf()
    plt.plot(x_coords, y_coords)
    plt.axis("equal")
    plt.title(f":  {iterations+1}")
    plt.show()
    plt.draw()


axiom = "F-F-F-F-F"  
rule = "F-F++F+F-F-F"  
angle = 72  #   (2pi/5)
step = 5  

#    
for iterations in range(6):  
    draw_lsystem(axiom, rule, angle, iterations, step)
```
